<div class="alert alert-block alert-info">
Creating ELO Scores <br>
ELO is a mathmatical way of ranking teams and their performance.<br>
For the indpenedent variable K I used https://fivethirtyeight.com/features/how-we-calculate-nba-elo-ratings/ for reference. 
They are a website that tracks ELO historically.<br>
If a team wins then their ELO increases with margin of victory, difference in ELO, and away or home.
Throughout the season the ELO accrues. 
Once a season end there is  a slight reset of ELO weighed in with both the baseline (1500 is the average ELO) and the teams last season. <br>
</div>

```python
TS_df['team'].unique()
```

```python
TS_df['ELO'] = 1500
TS_df['ELO_opp'] =1500
```

```python
Scores = {'ATL':1500, 'CHI':1500, 'DET':1500, 'CLE':1500, 'NOP':1500, 'GSW':1500, 'WAS':1500, 'ORL':1500, 'TOR':1500,
       'IND':1500, 'MIA':1500, 'BOS':1500, 'PHI':1500, 'CHO':1500, 'UTA':1500, 'BRK':1500, 'OKC':1500, 'MEM': 1500,
       'NYK':1500, 'MIL':1500, 'SAS':1500, 'DEN':1500, 'HOU':1500, 'DAL':1500, 'LAC':1500, 'SAC':1500, 'PHO':1500,
       'POR':1500, 'MIN':1500, 'LAL':1500}
```

```python
def getELO(won,margin,old,old_opp,home):
    if won == 1:
        K= (margin+3)**.8
        K = 7.5+(.006*(old -old_opp))
        K=K*20
        if home == 1:
            exp =(old_opp-(old+100))/400
            prob_win =1/(1+10**exp)
            new = old+(K*(1-prob_win))
        else:
            exp =(old_opp-old+100)/400
            prob_win =1/(1+10**exp)
            new = old+K*((1-prob_win))
    else:
        K= (margin+3)**.8
        K = 7.5+(.006*(old_opp-old))
        K=K*20
        if home == 1:
            exp =(old_opp-(old+100))/400
            prob_win =1/(1+10**exp)
            new = old+K*(0-prob_win)
        else:
            exp =(old_opp-old+100)/400
            prob_win =1/(1+10**exp)
            new = old+K*(0-prob_win)
    return new
        
```

```python
curSea = 2016
for index, game in TS_df.iterrows():
    if game['Season'] != curSea:
        for team in Scores.keys():
            Scores[team] = (1500*.25)+(Scores[team]*.75)
    Scores[game['team']] = getELO(game['won'],np.abs(game['total']-game['total_opp']),Scores[game['team']],Scores[game['team_opp']],game['home'])
    TS_df['ELO'].loc[index]= Scores[game['team']]
    TS_df['ELO_opp'].loc[index] = Scores[game['team_opp']]
```

<h2 align="center"> <font color='grey'>Feature Selection</font></h2>
<div>
1. Eliminating through coefficient correlation <br>
2. Normalization <br>
3. Creating rolling averages to the points and averages to the last dates<br>
4. Using backwards feature selection<br>
</div>

The upper tri function here is to avoid removing the diagonal

```python
cor_matrix = TS_df.corr().abs()
upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(bool))
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.95)]
TS_df2 = TS_df.drop(to_drop, axis=1)
```

```python
print('column names: \n')
print('\n'.join(list(TS_df2.columns)))
print(TS_df2.shape)
```

Normalizing with a MinMax

```python
dkeep=['Season','team','won','team_opp','home']
keep = TS_df2.columns[~TS_df2.columns.isin(dkeep)]
scaler = MinMaxScaler()
scaler.fit(TS_df2[keep])
TS_df2[keep]= scaler.transform(TS_df2[keep])
```

Creating rolling averages
A rolling average is taking the average of n points before a certian period.
I decided for a rolling average of 10. THese means the model will not be able to predict for the first 10 games of each team.

```python
def getAvg10(team):
    return team.rolling(10).mean()        
```

```python
rolling10=TS_df2[list(keep)+['Season','team','won']]
rolling10 = rolling10.groupby(['team','Season'],group_keys=False).apply(getAvg10)
```

```python
rolling10.columns = [f'{col}_10' for col in rolling10.columns]
df3 = TS_df2[dkeep]
Alldf = pd.concat([rolling10,df3], axis=1).dropna()
```

Merging with the opposite teams stats and playability

```python
BothTeam = Alldf.merge(Alldf,left_on=['team','date'],right_on=['team_opp','date'],how='left').dropna()
```

```python
print('column names: \n')
print('\n'.join(list(BothTeam.columns)))
print(BothTeam.shape)
```

```python
BothTeam.drop(['Season_y','won_y','Season_10_y','Season_10_x'],axis=1,inplace=True)
BothTeam.sort_index(inplace=True)
```

Using Sequential Feature Selector to get best features for Logistical Regression Model

Starting With Logistic regression there is one more transformation I want to apply to the data. This is PCA

```python
y_train, y_test =None,None
X_train, X_test = None,None
X= X[saved_features]
from sklearn.model_selection import TimeSeriesSplit
tss = TimeSeriesSplit(n_splits = 3)
for train_index, test_index in tss.split(X):
    X_train, X_test = X.iloc[train_index, :], X.iloc[test_index,:]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
```

```python
from sklearn.metrics import accuracy_score
logreg=LogisticRegression(class_weight='balanced',solver='lbfgs',random_state=42,n_jobs=-1,max_iter=500)
logreg.fit(X_train,y_train)
y_pred = logreg.predict(X_train)
train_acc = accuracy_score(y_train,y_pred)
print("Train accuracy :",train_acc)
y_pred = logreg.predict(X_test)
test_acc = accuracy_score(y_test,y_pred)
print("Testing accuracy :",test_acc)
```

For PCA we want to reduce our dimensionality by one third

```python
train_pca = pca.transform(X_train)
test_pca = pca.transform(X_test)
```

```python
logreg.fit(train_pca,y_train)
y_pred = logreg.predict(train_pca)
train_acc = accuracy_score(y_train,y_pred)
print("Train accuracy :",train_acc)
y_pred = logreg.predict(test_pca)
test_acc = accuracy_score(y_test,y_pred)
print("Testing accuracy :",test_acc)
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import csv
```

Read in data pulled from Spider into a Data frame.
Changing the seasons attribute of df to represent season rather than urls

```python
df =pd.read_csv('BoxScoreLinks.csv')
seasons = pd.Series(df['Season'])
new_names = [seasons[i][49:len(seasons[i])-5] for i in range(len(seasons))]
df =df.drop('Season', axis=1)
df['Season']= new_names
df =df.sort_values(by='Season').reset_index()
print(df['Season'][0:9])
```

Defining a function to pull each boxscore link out of each season

```python
def getBoxScore(boxscore,out_filename):
    all_links = []
    for i in boxscore:
        soup = BeautifulSoup(i,'lxml')
        links =soup.find_all('a')
        href = [a.get('href') for a in links]
        box_scores = [link for link in href if link  and "boxscore" in link and ".html" in link]
        box_scores = [f"https://www.basketball-reference.com{link}" for link in box_scores]
        all_links.append(box_scores)
    with open(out_filename, 'w', newline='') as out_file:
        fields = ['URL']
        write =csv.DictWriter(out_file,fieldnames=fields)
        for i in all_links:
            for j in i:
                dic = {fields[0] : j}
                write.writerow(dic)
```
